**Simple Linear Regression**

```python
import numpy as np
from sklearn import datasets, linear_model
```

```python
x,y,coef=datasets.make_regression(n_samples=100,
                                  n_features=1,
                                  n_informative=1,
                                  noise=10,
                                  coef=True,
                                  random_state=0)
```

```python
# Scale the feature x for exp from 0 to 20
x=np.interp(x,(x.min(),x.max()),(0,20))
```

```python
len(x)
```

```python
print(x)
```

```python
y=np.interp(y,(y.min(),y.max()),(20000,150000))
len(y)
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'.',label='Training Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_model=LinearRegression()
reg_model.fit(x,y)
```

```python
y_predicted=reg_model.predict(x)
```

```python
plt.plot(x,y_predicted,'.',color="black",label='Predicted Values')
plt.plot(x,y,'.',label="Actual Values")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.legend()
```

```python
# Creating regression model object
reg_model=LinearRegression()
x=np.random.rand(100,1)
error = np.random.rand(100,1)
b0 = 10
b1 = 7
y = b0+b1*x+error
```

```python
reg_model.fit(x,y)
y_predicted=reg_model.predict(x)
```

```python
plt.plot(x,y_predicted,".",color="black",label="Predicted Data")
plt.plot(x,y,".",label="Actual Data")
plt.xlabel("x")
plt.ylabel("y")
plt.title("X vs Y")
plt.legend()
```
