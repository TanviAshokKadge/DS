```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
from urllib.request import urlopen

# HTML Scraping
url_html = 'https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'
response = requests.get(url_html, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', class_='wikitable')
rows = table.find_all('tr')[1:]

countries, areas = [], []
for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:
        countries.append(cells[1].get_text(strip=True))
        areas.append(cells[3].get_text(strip=True).replace(',', ''))

df_html = pd.DataFrame({'Country': countries, 'Area': areas})
print('HTML Scraping Header:\n', df_html.head())

# JSON Scraping
url_json = 'https://jsonplaceholder.typicode.com/users'
data = json.loads(urlopen(url_json).read())
df_json = pd.DataFrame(data)[['id', 'username', 'email']]
print('\nJSON Scraping Header:\n', df_json.head())
```
