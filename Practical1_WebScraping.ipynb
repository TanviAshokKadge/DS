## HTML Scraping Code

```
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')
SrNo, Country, Area = [], [], []

rows = table.find("tbody").find_all("tr")
for row in rows:
    cells = row.find_all('td')
    if cells:
        SrNo.append(cells[0].get_text().strip())
        Country.append(cells[1].get_text().strip())
        Area.append(cells[3].get_text().strip().replace(",", ""))

df = pd.DataFrame({"SrNo": SrNo, "Country": Country, "Area": Area})
df.head()
```

## JSON Scraping Code

```
import json
from urllib.request import urlopen
import pandas as pd

url = "https://jsonplaceholder.typicode.com/users"
data = json.loads(urlopen(url).read())

Id, Username, Email = [], [], []

for item in data:
    Id.append(item.get("id","NA"))
    Username.append(item.get("username","NA"))
    Email.append(item.get("email","NA"))

df = pd.DataFrame({"Id": Id, "Username": Username, "Email": Email})
df.head()
```
