```
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')
SrNo, Country, Area = [], [], []

rows = table.find("tbody").find_all("tr")
for row in rows:
    cells = row.find_all('td')
    if cells:
        SrNo.append(cells[0].get_text().strip())
        Country.append(cells[1].get_text().strip())
        Area.append(cells[3].get_text().strip().replace(",", ""))

df = pd.DataFrame({"SrNo": SrNo, "Country": Country, "Area": Area})
df.head()
```
