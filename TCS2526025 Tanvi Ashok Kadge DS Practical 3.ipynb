```python
# 'axis = 1' -> columns & 'axis = 0' -> rows 

mtcars_cleaned = mtcars_df.drop(['model', 'carb'], axis=1)
mtcars_cleaned.head()
```

```python
mtcars_cleaned.info()
```

```python
from matplotlib import pyplot as plt 
import seaborn as sns  

sns.catplot(x="am", hue="cyl", kind="count", data=mtcars_cleaned)
```

```python
mtcars_df.isnull().sum()
```

```python
mtcars_cleaned.groupby(["am","cyl"])["cyl"].count()
```

```python
group1 = mtcars_cleaned.groupby(["am","cyl"])
trans_cyl = group1.size().unstack()
sns.heatmap(trans_cyl, annot=True, fmt="d")
```

```python
sns.violinplot(
    x="am",
    y="mpg",
    hue="vs",
    data=mtcars_cleaned,
    split=True,
)
```

```python
mtcars_corr = mtcars_cleaned.corr(method="pearson")
mtcars_corr
```

```python
sns.heatmap(mtcars_corr, annot=True, vmax=1)
```
